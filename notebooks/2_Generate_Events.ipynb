```python
import ast
import random
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('p_df.csv')
```

```python
p_df = df.drop('Unnamed: 0', axis=1)
```

```python
p_df
```

```python
p_df['photo'] = p_df['photo'].apply(lambda x: 'https://scrii.github.io/evops-dummy-photos/' + x)
```

```python
def truncate_title(title, max_chars=61):
    if len(title) <= max_chars:
        return title
    truncated = title[:max_chars]
    last_space_index = truncated.rfind(' ')
    if last_space_index == -1:
        return truncated + '...'
    else:
        return truncated[:last_space_index] + '...'
```

```python
p_df["preprocessed_text"] = p_df['text'].apply(lambda x: x.replace('\n', '\\n'))
p_df["preprocessed_text"] = p_df['preprocessed_text'].apply(lambda x: x.replace('\'', "\\'"))
p_df["preprocessed_text"] = p_df['preprocessed_text'].apply(lambda x: x.replace('\"', '\\"'))

p_df["preprocessed_titles"] = p_df['titles'].apply(lambda x: x.replace('\n', '\\n'))
p_df["preprocessed_titles"] = p_df['preprocessed_titles'].apply(lambda x: x.replace('\'', "\\'"))
p_df["preprocessed_titles"] = p_df['preprocessed_titles'].apply(lambda x: x.replace('\"', '\\"'))

p_df['truncated_title'] = p_df['preprocessed_titles'].apply(lambda x: truncate_title(x))
```

```python
photos = []

for idx, row in p_df.iterrows():
    if idx >= 312:
        photos.append(row['photo'])

cpy = photos.copy()
```

```python
def get_pics(amount: int):
    choose = random.sample(cpy, amount)
    for im in choose:
        cpy.remove(im)
    return ", ".join(list(map(lambda x: "\"" + x + "\"", choose)))
```

```python
for idx, row in p_df.iterrows():
    if idx > 199:
        break
    t_tags = ast.literal_eval(row['tags'])
    t_tags = ', '.join(list(map(lambda x: 'tags[\"' + '-'.join(x.lower().split()) + '\"].id', t_tags)))
    desc = row['preprocessed_text']
    attendance = str(random.choice([False, True])).lower()
    photo_link = row['photo']
    
    r = np.random.choice(5)
    if r == 0:
        photo_link = ""
    else:
        photo_link = get_pics(r)

    print(
f'''
  await createEvent(
    api,
    {{
      authorId: users.o4u.id,
      title: \"{row['truncated_title']}\",
      description:
        \"{desc}\",
      tagIds: [{t_tags}],
      withAttendance: {attendance},
    }},
    [{photo_link}],
  );
''',end='')
```
